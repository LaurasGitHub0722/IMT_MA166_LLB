```python
# Letzte Zeile einer txt datei auslesen und in csv speichern

import pandas as pd
import pathlib
from pathlib import Path

number = 10000
Nummer = []
AnzahlFramesPhase = []
AnzahlFramesTool = []
AnzahlFrames_data_resize = []
AnzahlFramesVideo = []

while number < 390601: 
    phasefile = Path('/content/drive/MyDrive/MA_LL/Datensatz_1143/Phase_annotations_1143/Greifen_1_%d-phase.txt' % number)
    if phasefile.is_file():
      with open(phasefile, 'r') as f:
          last_line_phase = f.readlines()[-1]
      AnzahlFramesPhase.append(last_line_phase)
      Nummer.append(number)
      print(number)    
      print('Letzte Zeile Label: {}'.format(last_line_phase))
      #toolfile = Path('/content/drive/MyDrive/MA_LL/Tool_annotations_333/Greifen_1_%d-tool.txt' % number)
      #with open(toolfile, 'r') as }f:
      #    last_line_tool = f.readlines()[-1]
      #AnzahlFramesTool.append(last_line_tool)
      #print(last_line_tool)

      framezahl = 0
      for path in pathlib.Path("/content/drive/MyDrive/MA_LL/Datensatz_1143/data_resize/Greifen_1_%d" % number).iterdir():
        if path.is_file():
          framezahl += 1
      AnzahlFrames_data_resize.append(framezahl)
      print('data_resize Framezahl: {}'.format(framezahl))
      
      framezahl_frame = 0
      for path in pathlib.Path("/content/drive/MyDrive/MA_LL/Datensatz_1143/Frame_Ordner/Greifen_1_%d" % number).iterdir():
        if path.is_file():
          framezahl_frame += 1
      AnzahlFramesVideo.append(framezahl_frame)
      print('Frame_Ordner Framezahl: {}'.format(framezahl_frame))
      number += 1
    else:
      number += 1

print('Done')
print('')

phase_video_df = pd.DataFrame(columns=["Nummer","AnzahlFramesPhase", 'AnzahlFrames_data_resize', 'AnzahlFramesVideo']) #'AnzahlFramesTool'
phase_video_df["Nummer"] = Nummer
phase_video_df['AnzahlFramesPhase'] =  AnzahlFramesPhase
#phase_video_df['AnzahlFramesTool'] =  AnzahlFramesTool
phase_video_df['AnzahlFrames_data_resize'] =  AnzahlFrames_data_resize
phase_video_df['AnzahlFramesVideo'] = AnzahlFramesVideo


phase_video_df.to_csv (r'/content/drive/MyDrive/MA_LL/Datensatz_1143/phase_video_csv_.csv', index = False, header=True)
```
